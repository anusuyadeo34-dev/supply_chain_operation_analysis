```python
# ============================================================
# IMPORT LIBRARIES
# ============================================================

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

sns.set_style("whitegrid")

print("Libraries Imported Successfully")
```

```python
import os

print(os.listdir())
```

```python
# ============================================================
# LOAD DATASET
# ============================================================

df = pd.read_csv("APL_Logistics.csv", encoding='latin1')

print("Dataset Loaded Successfully")
```

```python
# ============================================================
# FIRST 5 ROWS
# ============================================================

display(df.head())
```

```python
# ============================================================
# DATASET SHAPE
# ============================================================

print("Dataset Shape :", df.shape)
```

```python
# ============================================================
# COLUMN NAMES
# ============================================================

for column in df.columns:
    print(column)
```

```python
# ============================================================
# DATASET INFORMATION
# ============================================================

df.info()
```

```python
# ============================================================
# STATISTICAL SUMMARY
# ============================================================

display(df.describe())
```

```python
# ============================================================
# MISSING VALUES
# ============================================================

print(df.isnull().sum())
```

```python
# ============================================================
# DUPLICATE VALUES
# ============================================================

print("Duplicate Values :", df.duplicated().sum())
```

```python
# ============================================================
# REMOVE DUPLICATES
# ============================================================

df.drop_duplicates(inplace=True)

print("Duplicates Removed Successfully")
```

```python
# ============================================================
# HANDLE MISSING VALUES
# ============================================================

df.fillna(0, inplace=True)

print("Missing Values Handled")
```

```python
from pandas.core.groupby.generic import DataFrameGroupBy
# ============================================================
# KPI ANALYSIS
# ============================================================

total_revenue = df['Sales'].sum()

total_profit = df['Order Profit Per Order'].sum()

profit_margin = (total_profit / total_revenue) * 100

total_orders = len(df)

print("Total Revenue :", round(total_revenue, 2))

print("Total Profit :", round(total_profit, 2))

print("Profit Margin :", round(profit_margin, 2), "%")

print("Total Orders :", total_orders)
```

```python
# ============================================================
# REVENUE TREND ANALYSIS
# ============================================================

plt.figure(figsize=(12,6))

plt.plot(df['Sales'])

plt.title("Revenue Trend Analysis")

plt.xlabel("Index")

plt.ylabel("Sales")

plt.tight_layout()

plt.savefig("revenue_trend.png")

plt.show()
```

```python
# ============================================================
# SALES DISTRIBUTION
# ============================================================

plt.figure(figsize=(10,5))

sns.histplot(df['Sales'], bins=30)

plt.title("Sales Distribution")

plt.xlabel("Sales")

plt.ylabel("Frequency")

plt.tight_layout()

plt.savefig("sales_distribution.png")

plt.show()
```

```python
# ============================================================
# PROFIT TREND ANALYSIS
# ============================================================

plt.figure(figsize=(12,6))

plt.plot(df['Order Profit Per Order'])

plt.title("Profit Trend Analysis")

plt.xlabel("Index")

plt.ylabel("Profit")

plt.tight_layout()

plt.savefig("profit_trend.png")

plt.show()
```

```python
# ============================================================
# PROFIT DISTRIBUTION
# ============================================================

plt.figure(figsize=(10,5))

sns.histplot(df['Order Profit Per Order'], bins=30)

plt.title("Profit Distribution")

plt.xlabel("Profit")

plt.ylabel("Frequency")

plt.tight_layout()

plt.savefig("profit_distribution.png")

plt.show()
```

```python
# ============================================================
# CATEGORY PROFITABILITY ANALYSIS
# ============================================================

category_profit = df.groupby('Category Name')[
    'Order Profit Per Order'
].sum().sort_values(ascending=False)

print(category_profit)
```

```python
# ============================================================
# CATEGORY PROFITABILITY CHART
# ============================================================

plt.figure(figsize=(10,5))

category_profit.plot(kind='bar')

plt.title("Category Wise Profitability")

plt.xlabel("Category")

plt.ylabel("Profit")

plt.tight_layout()

plt.savefig("category_profitability.png")

plt.show()
```

```python
# ============================================================
# TOP PRODUCTS
# ============================================================

product_profit = df.groupby('Product Name')[
    'Order Profit Per Order'
].sum().sort_values(ascending=False)

print(product_profit.head(10))
```

```python
# ============================================================
# TOP PRODUCTS CHART
# ============================================================

plt.figure(figsize=(12,6))

product_profit.head(10).plot(kind='bar')

plt.title("Top 10 Profitable Products")

plt.xlabel("Product")

plt.ylabel("Profit")

plt.tight_layout()

plt.savefig("top_products.png")

plt.show()
```

```python
# =========================================================
# CREATE CUSTOMER FULL NAME
# =========================================================

df['Customer Full Name'] = (
    df['Customer Fname'].astype(str)
    + " " +
    df['Customer Lname'].astype(str)
)

print(df['Customer Full Name'].head())
```

```python
# ============================================================
# TOP CUSTOMERS
# ============================================================

customer_profit = df.groupby('Customer Full Name')[
    'Order Profit Per Order'
].sum().sort_values(ascending=False)

print(customer_profit.head(10))
```

```python
# ============================================================
# TOP CUSTOMERS CHART
# ============================================================

plt.figure(figsize=(12,6))

customer_profit.head(10).plot(kind='bar')

plt.title("Top 10 Customers by Profit")

plt.xlabel("Customer")

plt.ylabel("Profit")

plt.tight_layout()

plt.savefig("top_customers.png")

plt.show()
```

```python
# ============================================================
# CUSTOMER SEGMENT ANALYSIS
# ============================================================

segment_profit = df.groupby('Customer Segment')[
    'Order Profit Per Order'
].sum()

print(segment_profit)
```

```python
# ============================================================
# CUSTOMER SEGMENT CHART
# ============================================================

plt.figure(figsize=(8,5))

segment_profit.plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title("Customer Segment Contribution")

plt.ylabel("")

plt.tight_layout()

plt.savefig("customer_segment.png")

plt.show()
```

```python
# ============================================================
# DISCOUNT VS PROFIT
# ============================================================

plt.figure(figsize=(10,6))

sns.scatterplot(
    x=df['Order Item Discount Rate'],
    y=df['Order Profit Per Order']
)

plt.title("Discount vs Profit Analysis")

plt.xlabel("Discount Rate")

plt.ylabel("Profit")

plt.tight_layout()

plt.savefig("discount_vs_profit.png")

plt.show()
```

```python
# ============================================================
# CORRELATION ANALYSIS
# ============================================================

correlation = df[
    ['Order Item Discount Rate',
     'Order Profit Per Order']
].corr()

print(correlation)
```

```python
# ============================================================
# REGION WISE PROFIT
# ============================================================

region_profit = df.groupby('Order Region')[
    'Order Profit Per Order'
].sum().sort_values(ascending=False)

print(region_profit)
```

```python
# ============================================================
# REGION PROFIT CHART
# ============================================================

plt.figure(figsize=(12,6))

region_profit.plot(kind='bar')

plt.title("Regional Profitability Analysis")

plt.xlabel("Region")

plt.ylabel("Profit")

plt.tight_layout()

plt.savefig("regional_profitability.png")

plt.show()
```

```python
# ============================================================
# COUNTRY SALES ANALYSIS
# ============================================================

country_sales = df.groupby('Order Country')[
    'Sales'
].sum().sort_values(ascending=False).head(10)

print(country_sales)
```

```python
# ============================================================
# COUNTRY SALES CHART
# ============================================================

plt.figure(figsize=(12,6))

country_sales.plot(kind='bar')

plt.title("Top Countries by Revenue")

plt.xlabel("Country")

plt.ylabel("Sales")

plt.tight_layout()

plt.savefig("top_countries.png")

plt.show()
```

```python
# ============================================================
# HEATMAP ANALYSIS
# ============================================================

numeric_columns = df.select_dtypes(include=np.number)

correlation_matrix = numeric_columns.corr()

plt.figure(figsize=(14,8))

sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")

plt.tight_layout()

plt.savefig("correlation_heatmap.png")

plt.show()
```
